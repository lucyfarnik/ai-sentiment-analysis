```python
import os
import pandas as pd
from tqdm import tqdm

def extract_country():
    file_path = 'data/merged_data_sentiment.csv'
    df = pd.read_csv(file_path)
#     print(df)
#     print(df.iloc[9205])
    print(df.iloc[29721])
    nanCounter = 0
    platform_counts = {'Twitter': {'total': 0, 'nans': 0},
                       'youtube': {'total': 0, 'nans': 0},
                       'Reddit': {'total': 0, 'nans': 0}}
    for i, row in tqdm(df.iterrows()):
        platform = row['Platform']
        if pd.isna(platform):
            print(i)
#             print(platform)
            continue
        platform_counts[platform]['total'] += 1
        if pd.isna(row['Country']):
            nanCounter += 1
            platform_counts[platform]['nans'] += 1

    print(f"Total Nans: {nanCounter} out of {len(df)} items")
    for platform, counts in platform_counts.items():
        total = counts['total']
        nans = counts['nans']
        print(f"{platform} Nans: {nans} out of {total} items")

if __name__ == '__main__':
    extract_country()
```
