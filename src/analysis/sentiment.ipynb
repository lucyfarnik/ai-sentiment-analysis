```python
import pandas as pd
import transformers
```

```python
df = pd.read_csv('../../data/yt_w_langs_en.csv')
```

```python
classifiers = [transformers.pipeline('sentiment-analysis', model) for model in [
   'distilbert-base-uncased-finetuned-sst-2-english',
   'cardiffnlp/twitter-roberta-base-sentiment',
   'finiteautomata/bertweet-base-sentiment-analysis',
   'j-hartmann/emotion-english-distilroberta-base',
]]

classify = lambda inpt: [cls(inpt[:400]) for cls in classifiers]
```

```python
geodata = df[df['country'].notna()]
geodata
```

```python
classify(geodata['text'][20:30].to_list())
```
